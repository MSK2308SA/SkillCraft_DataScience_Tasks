```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv('marketing_campaign1.csv', sep=';')
print("Shape:", df.shape)
print(df.head())

# Clean data
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'])
print("Cleaned! Rows remaining:", len(df))

# Chart 1 - Purchases by Channel
channels = ['NumWebPurchases', 'NumCatalogPurchases',
            'NumStorePurchases', 'NumDealsPurchases']
channel_totals = df[channels].sum()

plt.figure(figsize=(8,5))
channel_totals.plot(kind='bar', color=['blue','green','orange','red'])
plt.title('Total Purchases by Channel')
plt.xlabel('Channel')
plt.ylabel('Number of Purchases')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Chart 2 - Spend by Product Category
spend_cols = ['MntWines', 'MntFruits',
              'MntMeatProducts', 'MntFishProducts',
              'MntSweetProducts', 'MntGoldProds']
spend_totals = df[spend_cols].sum()

plt.figure(figsize=(8,5))
spend_totals.plot(kind='bar', color='purple')
plt.title('Total Spend by Product Category')
plt.xlabel('Category')
plt.ylabel('Total Amount Spent')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Chart 3 - Campaign Response Rate
campaigns = ['AcceptedCmp1','AcceptedCmp2','AcceptedCmp3',
             'AcceptedCmp4','AcceptedCmp5']
response_rate = df[campaigns].sum()

plt.figure(figsize=(8,5))
response_rate.plot(kind='bar', color='teal')
plt.title('Campaign Response Rate')
plt.xlabel('Campaign')
plt.ylabel('Number of Responses')
plt.tight_layout()
plt.show()

print("All charts generated successfully!")
```
